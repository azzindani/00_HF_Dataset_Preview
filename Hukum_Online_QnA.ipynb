```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.hukumonline.com/klinik/arsip/'

response = requests.get(url)
```

```python
if response.status_code == 200:
  soup = BeautifulSoup(response.content, 'html.parser')
  article_titles = soup.find_all(class_='klinik-list d-flex flex-column divider-bottom my-3')

  for index, title in enumerate(article_titles):
    link = title.find('a')

    if link:
      article_url = 'https://www.hukumonline.com' + link['href']
      article_text = link.get_text()

      print(f"Title: {article_text}")
      print(f"Link: {article_url}")

      if index == 1:
        article_response = requests.get(article_url)

        if article_response.status_code == 200:
          article_soup = BeautifulSoup(article_response.content, 'html.parser')

          print("\nExtracted Text from the Article:\n")
          article_content = article_soup.find_all('p')
          for paragraph in article_content:
            print(paragraph.get_text())

          print("\nExtracted Tables from the Article:\n")
          tables = article_soup.find_all('table')

          for table in tables:
            rows = table.find_all('tr')

            for row in rows:
              cells = row.find_all(['td', 'th'])
              cell_text = [cell.get_text(strip=True) for cell in cells]

              print("\t".join(cell_text))

        else:
            print(f"Failed to open the article page. Status code: {article_response.status_code}")
else:
  print(f"Failed to retrieve the page. Status code: {response.status_code}")
```

```python
if response.status_code == 200:
  soup = BeautifulSoup(response.content, 'html.parser')
  article_titles = soup.find_all(class_='klinik-list d-flex flex-column divider-bottom my-3')

  for index, title in enumerate(article_titles):
    link = title.find('a')

    if link:
      article_url = 'https://www.hukumonline.com' + link['href']
      article_text = link.get_text()

      print(f"Title: {article_text}")
      print(f"Link: {article_url}")

      if index == 1:
        if response.status_code == 200:
          soup = BeautifulSoup(response.content, 'html.parser')
          start_header = soup.find('h2', string=re.compile(r'PERTANYAAN', re.I))
          end_header = soup.find('h2', string=re.compile(r'TAGS', re.I))

          if start_header is None:
            print("Start header 'PERTANYAAN' not found.")
          if end_header is None:
            print("End header 'TAGS' not found.")

          if start_header and end_header:
            content_between = []
            current_element = start_header.find_next_sibling()

            while current_element and current_element != end_header:
              if current_element.name:
                content_between.append(current_element.get_text(strip=True))
              current_element = current_element.find_next_sibling()
          print("\nText between 'PERTANYAAN' and 'TAGS':\n")
          print("\n".join(content_between))
        else:
            print("Could not find the required headers.")

else:
  print(f"Failed to retrieve the page. Status code: {response.status_code}")
```
